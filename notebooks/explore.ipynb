Plot loss and forecast for runs

```python
import seaborn as sns
import matplotlib.pyplot as plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dyna_loss_fp = 'stable_baselines3/dyna_loss.csv'
forecast_fp = 'stable_baselines3/forecast.csv'

def set_second_xlabel(ax, positions, labels):
    ax2 = ax.twiny()
    ax2.set_xlim(ax.get_xlim())
    ax2.set_xticks(positions)
    ax2.set_xticklabels(labels)
    return ax2
```

```python
dyna_loss = pd.read_csv(dyna_loss_fp, names=['step', 'episode', 'loss'])
forecasts = pd.read_csv(forecast_fp, names=['step', 'episode', 'forecast'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
dyna_loss.set_index('step').loss.rolling(200).mean().plot(ax=ax)
ax2 = set_second_xlabel(ax, dyna_loss.step[::4000], dyna_loss.episode[199::4000])
ax.set_ylabel('loss')
ax2.set_xlabel('episode')
plt.title('Dynamics Model Loss', fontsize=20)
```

```python

fig = plt.figure()
ax = fig.add_subplot(111)
forecasts.set_index('step').forecast.rolling(200).mean().plot(ax=ax)
ax2 = set_second_xlabel(ax, forecasts.step[199::100], forecasts.episode[199::100])
ax.set_ylabel('forecast')
ax2.set_xlabel('episode')
plt.title('Forecast', fontsize=20)
```
